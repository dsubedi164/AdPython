# List
    list is an ordered, mutable collection of items. 
    It can contain elements of any data type, including other lists. 
    Lists are one of the most commonly used data structures in Python.

## 1. Creating Lists

### a. Empty List

```python
# Method 1
my_list = []
print(my_list)  

# Method 2
my_list = list()
print(my_list)  
```

### b. List with Elements


```python
# A list of integers
numbers = [1, 2, 3, 4, 5]
print(numbers)

# A list of strings
fruits = ["apple", "banana", "cherry"]
print(fruits)

# Mixed data types
mixed = [1, "apple", 3.14, True]
print(mixed)
```

### c. List from a Range

```python
# Using range() to create a list of numbers from 0 to 4
numbers = list(range(5))
print(numbers)  # Output: [0, 1, 2, 3, 4]

# Using range with start, stop, step
numbers = list(range(1, 11, 2))  # 1 to 10 with step 2
print(numbers)  # Output: [1, 3, 5, 7, 9]
```

### d. List from another iterable

```python
# Converting a string to a list of characters
chars = list("Python")
print(chars)  # Output: ['P', 'y', 't', 'h', 'o', 'n']

# Converting a tuple to a list
tuple_data = (1, 2, 3)
list_data = list(tuple_data)
print(list_data)  # Output: [1, 2, 3]
```

## 2. Accessing List Elements

```python
fruits = ["apple", "banana", "cherry", "date"]

print(fruits[0])   # First element -> apple
print(fruits[-1])  # Last element -> date
print(fruits[1:3]) # Slicing -> ['banana', 'cherry']
```

## 3. Modifying Lists

```python
fruits = ["apple", "banana", "cherry"]

# Changing an element
fruits[1] = "blueberry"
print(fruits)  # Output: ['apple', 'blueberry', 'cherry']

# Adding elements
fruits.append("date")
print(fruits)  # Output: ['apple', 'blueberry', 'cherry', 'date']

fruits.insert(1, "banana")
print(fruits)  # Output: ['apple', 'banana', 'blueberry', 'cherry', 'date']

# Removing elements
fruits.remove("cherry")
print(fruits)  # Output: ['apple', 'banana', 'blueberry', 'date']

# Pop element by index
removed = fruits.pop(2)
print(removed)  # Output: blueberry
print(fruits)   # Output: ['apple', 'banana', 'date']
# Pop last element
last = fruits.pop()
print(last)    # Output: date
print(fruits)  # Output: ['apple', 'banana']
```

## 4. List Operations

```python
a = [1, 2, 3]
b = [4, 5, 6]

# Concatenation
c = a + b
print(c)  # Output: [1, 2, 3, 4, 5, 6]

# Repetition
d = a * 2
print(d)  # Output: [1, 2, 3, 1, 2, 3]

# Membership
print(2 in a)   # Output: True
print(5 not in a) # Output: True

# Length 
print(len(a))  # Output: 3

# Iteration
for item in a:
    print(item)
                                # Output:
                                # 1
                                # 2
                                # 3       

# Nested Lists
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(matrix[1][2]) # Output: 6
```
