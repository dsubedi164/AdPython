# Functions and Lambda Expressions in Python

## 1. Functions in Python

A **function** is a reusable block of code that performs a specific task. Functions improve **modularity, readability, and reusability**.

### 1.1 Syntax of a Function
```python
def function_name(parameters):
    """Docstring explaining the function"""
    statements
    return value  # optional
```

````python
# Example: Basic Function
def greet(name):
    """Greets a person"""
    return f"Hello, {name}!"

print(greet("Python")) # Output: Hello, Python!
````

### 1.2 Types of Functions
| Type | Description | Example |
|------|------------|---------|
| No parameters, no return | A function that does not take any input and does not return any value. | `def say_hello(): print("Hello!")` |
| With parameters, no return | A function that takes input (parameters) but does not return a value. It typically performs an action using the inputs, like printing a customized message or saving data. | `def greet(name): print(f"Hi {name}")` |
| No parameters, with return | A function that doesn’t take any input but returns a value. The returned value can be stored or used in expressions. | `def get_pi(): return 3.1416` |
| With parameters and return | A function that takes input parameters and returns a value. | `def add(a,b): return a+b` |

### 1.3 Function Arguments
1. **Positional Arguments**

    Arguments that are passed to the function in the exact order of the parameters.


````python
def add(a, b):
    return a + b

print(add(3, 5))  # Positional argument
````

2. **Keyword Arguments**

    Arguments that are passed using the parameter name, so the order does not matter.

````python
print(add(b=10, a=5))  # Keyword argument
````

3. **Default Arguments**

    Parameters can have default values. If no argument is passed for that parameter, the default value is used.

````python
def greet(name="Guest"):
    return f"Hello, {name}!"

print(greet())        # Hello, Guest!
print(greet("Adk")) # Hello, Adk!
````

4. **Variable-Length Arguments**

    Sometimes we want to pass any number of arguments to a function.
- `*args` — multiple positional arguments (tuple)
- `**kwargs` — multiple keyword arguments (dict)

````python
def show(*args, **kwargs):
    print("Args:", args)
    print("Kwargs:", kwargs)

show(1, 2, 3, a=10, b=20)
````

### 1.4 Scope of Variables
- **Local variables**: Defined inside a function, accessible only there.
- **Global variables**: Defined outside, accessible everywhere.
- Use `global` keyword to modify a global variable inside a function.

````python
x = 5

def modify():
    global x
    x = 10

modify()
print(x)  # 10
````

### 1.5 Nested Functions

A nested function (also called an inner function) is a function defined inside another function.
Nested functions are useful for encapsulation, closures, and functional programming patterns.


````python
def outer():
    def inner():
        return "Hello from inner!"
    return inner()

print(outer())
````

## 2. Lambda Functions (Anonymous Functions)

A **lambda function** is a small anonymous function, often used for short tasks. It can have any number of arguments but only one expression, and it implicitly returns the result of that expression. **Lambda functions** are often used when you need a simple function for a short time, especially with functions like map(), filter(), and sorted().

### Syntax:
```python
lambda arguments: expression
```

````python
# Example: Lambda function
square = lambda x: x**2
print(square(5))  # 25
````

### Lambda with Multiple Arguments

Multiple arguments can be passed separated by commas. Only one expression is allowed. Returns the value of that expression automatically.

````python
add = lambda a, b: a + b
print(add(3, 7))  # 10
````

### Using Lambda with `map()`
**map()** applies a function to each item of an iterable.

````python
nums = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x**2, nums))
print(squares)
````

### Using Lambda with `filter()`

**filter()** selects elements from an iterable for which a function returns True.

````python
nums = [1, 2, 3, 4, 5]
evens = list(filter(lambda x: x % 2 == 0, nums))  #Only numbers divisible by 2 are selected.
print(evens)  # [2, 4]
````

### Using Lambda with `sorted()`

You can sort based on custom criteria using key.

````python
points = [(2,3), (1,4), (3,1)]

sorted_points = sorted(points, key=lambda x: x[1])
print(sorted_points)  # [(3, 1), (2, 3), (1, 4)]
````
